```python
import random
import tiktoken
```

```python
samples = random.choices(docs, k=5)
```

```python
def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
import json

platinum_data = [{'text': doc.get_text(), 
  'doc_id': doc.path.name.split('.')[0], 
  'n_tokens': num_tokens_from_string(doc.get_text(), 'cl100k_base')
 } for doc in docs]

platinum_text_prepared_path.write_text(json.dumps(platinum_data))
```
